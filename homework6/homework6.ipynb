```python
def is_prime(n):
    if n <2:
        return False
    for i in range (2,n):
        if n % i == 0:
            return False 
    return True 
```

```python
def rows_with_prime(arr):
    result = []

    for row in arr:
        for value in row:
            if is_prime(value):
                result.append(row)
                break
    return np.array(result)
print (rows_with_prime(arr))
```

```python
def checkerboard():
    board = np.zeros((8,8), dtype=int)

    for i in range(8):
        for j in range (8):
            if (i+j) % 2 ==0:
                board[i][j]=1
    return board
def reverse_checkerboard(): 
    board = np.zeros((8,8), dtype=int)

    for i in range(8):
        for j in range (8):
            if (i+j) %2 == 1:
                board [i][j]=1
    return board 

print (checkerboard())
print ()
print (reverse_checkerboard())
```

```python
def expansion(arr, spaces): 
    result = []

    for word in arr: 
        result.append((" "*spaces).join(word))

    return np.array(result)

universe= np.array(['galaxy','clusters'])

print(expansion(universe,1))
print(expansion(universe,2))
```

```python
def secondDimmest(stars): 
    result = []

    for row in stars: 
        sorted_row = np.sort(row)
        result.append(sorted_row[-2])

    return np.array(result)

np.random.seed(123)
stars = np.random.randint(500,2000,(5,5))

print (stars)
print (secondDimmest(stars))
    
```
