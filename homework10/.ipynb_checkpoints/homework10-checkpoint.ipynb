```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
from astropy import units as u 
from astropy import constants as const
```

```python
df = pd.read_csv("solar_system.csv")
print (df.shape)
```

```python
df = df.set_index("Attribute").T
df.index.name = "Planet"
df.reset_index(inplace=True)
df.columns.name = None

print (df.shape)
```

```python
# the before shape was 20 x 11 and the after shape was 10 x 21 
# first line makes the planets the rows, second row makes the row index, the third turns the index into a regular column, and the last one removes the attribute label from columns
#the new shape is 11x20 because the reset index adds the planet column making it 21 instead of 20 the 11th row was the attribute column which became the column headder after transposing hence why it is now 10. 
print(df.columns.tolist())
#columns with units 'Mass (10^24kg)', 'Diameter (km)', 'Density (kg/m^3)', 'Gravity (m/s^2)', 'Escape Velocity (km/s)', 'Rotation Period (hours)', 'Length of Day (hours)', 'Distance from Sun (10^6 km)', 'Perihelion (10^6 km)', 'Aphelion (10^6 km)', 'Orbital Period (days)', 'Orbital Velocity (km/s)', 'Orbital Inclination (deg)', 'Orbital Eccentricity', 'Obliquity to Orbit (deg)', 'Mean Temperature (C)', 'Surface Pressure (bars)', 'Number of Moons', 17 
#without units are Planet, 'Ring System?', 'Global Magnetic Field?' 3 

print (df.to_string())
```

```python
print (df.dtypes)

for col in df.columns:
    print (f"\n{col}:")
    print (df[col].apply(type))
```

```python
# they are all pretty much string values 
#strings cant be used when preforming math and or calucations because numpy and astropy need to use actual numbers
```

```python
for col in df.columns: 
    if col not in ["Planet","Ring System?", "Global Magnetic Field?"]:
        df[col]=pd.to_numeric(df[col], errors="coerce")

print (df.dtypes)
```

```python
""" 1. now we see obejects, float 64 and int 64. 
2. Float64 is decimal numbers, objects are a mixture of strings for our unitless values, and int64 are very large whole number values.
3. the unitless columns are still object because they contain text not numbers"""
```

```python
def attach_units(df, old_col, new_col, unit): 
    new_values = []
    for val in df[old_col]: 
        new_values.append(val * unit)
    df[old_col]=new_values
    df.rename(columns={old_col:new_col}, inplace=True)
    return df 

df = attach_units(df, "Mass (10^24kg)", "Mass (kg)", 1e24 * u.kg)
print(df["Mass (kg)"])
```

```python
print (df.columns.tolist())

df_col_index = df.columns.get_loc("Aphelion (10^6 km)")
df.insert(df_col_index +1, "Semi-Major Axis (10^6 km)", (df["Perihelion (10^6 km)"] + df["Aphelion (10^6 km)"]) /2 * u.km * 1e6)

print (df.columns.tolist())
print (df["Semi-Major Axis (10^6 km)"])
```

```python
days_per_year = 365.25 * u.day / u.year 
df["Orbital Period (years)"] = df["Orbital Period (days)"] * u.day.to(u.year)
df.drop(columns=["Orbital Period (days)"], inplace=True) 

print (df.columns.tolist())
print (df["Orbital Period (years)"])
```

```python
planet = "Earth" 
val = df.loc[df["Planet"] == planet, "Orbital Period (years)"].values[0]
print (f"{planet}'s orbital period: {val:.4f} years")
```

```python
print (const.au)
print (const.au.to(u.km))

distance_cols ={"Diameter (km)":"Diameter (AU)", 
                "Perihelion (10^6 km)": "Perihelion (AU)", 
                "Aphelion (10^6 km)": "Aphelion (AU)", 
                "Semi-Major Axis (10^6 km)" : "Semi-Major Axis (AU)"}

km_per_au = const.au.to(u.km).value 

for old_col, new_col in distance_cols.items(): 
    if old_col in df.columns: 
        df[old_col] = df[old_col] / km_per_au
        df.rename (columns = {old_col:new_col}, inplace=True) 

for new_col in distance_cols.values(): 
    if new_col in df.columns:
        val = df.loc[df["Planet"] == "Earth", new_col].values[0]
        print (f"Earth's {new_col}: {val}")
```

```python
df.to_csv("units.csv", index=False)
```
