```python
log_magnitude = np.log(magnitude)
log_temperature = np.log(temperature)
 
```

```python
plt.figure(figsize=(8,6))

plt.scatter(log_temperature, log_magnitude, label="Stars")

plt.scatter(log_temperature[min_mag_index], log_magnitude[min_mag_index], color="black", marker="*", s = 300, label="brightest Star")

plt.scatter(log_temperature[max_mag_index], log_magnitude[max_mag_index], color="purple", marker="*", s = 300, label="Dimmest Star")

plt.scatter(log_temperature[max_temp_index], log_magnitude[max_temp_index], color="green", marker="*", s = 300, label="Hottest Star")

plt.scatter(log_temperature[min_temp_index], log_magnitude[min_temp_index], color="orange", marker="*", s = 300, label="Coldest Star")

plt.gca().invert_xaxis()
plt.gca().invert_yaxis()

plt.title("Hertzsprung-Russel Diagram")
plt.xlabel("Temperature (log K)")
plt.ylabel("magnitude")
plt.legend()

plt.show()
```
