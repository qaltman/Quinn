```python
import numpy as np 
import matplotlib.pyplot as plt

def make_sine_wave(x, A, w): 
    return A * np.sin(w *x)
    
```

```python
x= np.linspace(0, 2 * np.pi, 1000)
```

```python
amplitude = [0.5, 1, 1.5, 2, 2.5]
frequency = [1,2,3,4,5]
```

```python
fig, axes = plt.subplots(1,3, figsize=(25,5))

for A, w in zip(amplitude, frequency): 
    y1 = make_sine_wave(x,A,1)
    axes[0].plot(x,y1, label= f"A={A}, w=1")

    y2 = make_sine_wave(x,1,w)
    axes[1].plot(x,y2, label=f"A=1, w={w}")

    y3 = make_sine_wave(x, A, w)
    axes[2].plot(x,y3, label=f"A={A}, w={w}") 


axes[0].set_title("Varrying Amp", fontsize=15)
axes[0].set_xlabel("x", fontsize=13)
axes[0].set_ylabel("y= A sin(wx)", fontsize=13)
axes[0].legend(fontsize=10)

axes[1].set_title("Varrying Freq", fontsize=15)
axes[1].set_xlabel("x", fontsize=13)
axes[1].set_ylabel("y= A sin(wx)", fontsize=13)
axes[1].legend(fontsize=10)

axes[2].set_title("Varrying Amp and Freq", fontsize=15)
axes[2].set_xlabel("x", fontsize=13)
axes[2].set_ylabel("y= A sin(wx)", fontsize=13)
axes[2].legend(fontsize=10)

plt.show()
    
```
